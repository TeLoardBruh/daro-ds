```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# Sample dataset of movie descriptions and corresponding titles
descriptions = [
    "A young wizard's journey to defeat an evil sorcerer",
    "A group of friends embark on a quest to destroy a powerful ring",
    "A superhero protects the city from villains",
    "A heartwarming story about love and friendship",
    "A thrilling adventure set in outer space"
]

titles = [
    "Harry Potter",
    "The Lord of the Rings",
    "Spider-Man",
    "Love Actually",
    "Star Wars"
]

# Encode the movie titles as numerical labels
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(shuffled_df['business_industry'])

# Vectorize the movie descriptions using TF-IDF
vectorizer = TfidfVectorizer()
descriptions_vectorized = vectorizer.fit_transform(shuffled_df['sale_item'])

# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(descriptions_vectorized, encoded_labels, test_size=0.2, random_state=42)

# Create a neural network model
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(32, activation='relu'))
model.add(Dense(len(np.unique(encoded_labels)), activation='softmax'))

# Compile the model
model.compile(loss='sparse_categorical_crossentropy', optimizer=Adam(), metrics=['accuracy'])

# Train the model for a specified number of epochs
epochs = 10
model.fit(X_train.toarray(), y_train, epochs=epochs, batch_size=32, verbose=1)

# Evaluate the model on the test set
loss, accuracy = model.evaluate(X_test.toarray(), y_test)
print('Test Loss:', loss)
print('Test Accuracy:', accuracy)

# Guess the movie title for a new description
for i in range(len(data_test_set)):
  new_description_vectorized = vectorizer.transform([data_test_set[i]['itemPurchase']]).toarray()
  predicted_probabilities = model.predict(new_description_vectorized)
  predicted_label = np.argmax(predicted_probabilities)
  predicted_title = label_encoder.inverse_transform([predicted_label])
  print("Predicted Title:", predicted_title[0])
```
